```python
# 1. df.iloc[0]
# -------------------------

# iloc = integer-location based indexing.

# df.iloc[0] → accesses the first row (index position 0).

# Returns a Series representing that row.

# 2. df.iloc[4:6]
# -------------------------

# iloc[4:6] → accesses rows by integer positions 4 and 5 (Python slicing is exclusive of the last index).

# So it gives you the 5th and 6th rows.

# Returns a DataFrame with those two rows.

# 3. df.loc[4:6]
# -------------------------

# loc = label-based indexing.

# df.loc[4:6] → selects rows with labels from 4 to 6 (inclusive).

# -------------------------

# Important: Unlike iloc, loc slicing is inclusive of the last label.

# So if your DataFrame has default integer index starting from 0:

# iloc[4:6] → rows at positions 4 and 5

# loc[4:6] → rows with index labels 4, 5, and 6

# Example (assuming default integer index):


data = {
    "Name": ["Ankit", "Ravi", "Priya"],
    "Age": [21, 22, 23],
    "Course": ["BCA", "B.Tech", "MBA"]
}

df = pd.DataFrame(data)
print(df)


```
