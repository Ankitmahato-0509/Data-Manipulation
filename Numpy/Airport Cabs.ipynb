```python
import numpy as numpy
taxi_data = numpy.genfromtxt('nyc_taxis.csv', delimiter=',', names=True, dtype=None, encoding=None)
```

```python
speed = taxi_data['trip_distance'] / (taxi_data['trip_length'] / 3600)
print("Average speed:", numpy.mean(speed), "miles per hour")

```

```python
rides_february = taxi_data[taxi_data['pickup_month'] == 2]
print("Number of rides in February:", len(rides_february))
```

```python
tip_more_than_50 = taxi_data[taxi_data['tip_amount'] > 50]
print("Number of rides with tip more than $50:", len(tip_more_than_50))
```

```python
no_of_dropoffs_at_jfk_airport = taxi_data[taxi_data['dropoff_location_code'] == 2]
print("Number of dropoffs at JFK airport:", len(no_of_dropoffs_at_jfk_airport))
```
