```python
import numpy as np

data = np.genfromtxt(r"D:\ml\Project\NumPy\Icecream Sales wr Rain and Temperature.csv", delimiter=',', skip_header=1)

# Split into columns
# ------------------------------------------------------------------

temperatures = data[:, 0]       # all rows, first column (Temperature)
ice_cream_prices = data[:, 1]   # all rows, second column (Sales)
No_of_tourists = data[:, 2]     # all rows, third column (No of Tourists)
ice_cream_Sales = data[:, 3]     # all rows, fourth column (Ice Cream Sales)
is_raining = data[:, 4]        # all rows, fifth column (Is Raining)

print("Temperatures:", temperatures)
print("Ice Cream Prices:", ice_cream_prices)
print("No of Tourists(Thouands):", No_of_tourists)
print("Ice Cream Sales($):", ice_cream_Sales)
print("Is Raining:", is_raining)

print("======================================================\n")
# ------------------------------------------------------------------

print("Temperature(F)  Ice Cream Price($)  No of Tourists(Thousands)  Ice Cream Sales($)  Is Raining")
print("-" * 100)

for temp, price, tourists, sales, raining in zip(temperatures, ice_cream_prices, No_of_tourists, ice_cream_Sales, is_raining):
    print(f"{temp:15.1f}  {price:18.2f}  {tourists:24.1f}  {sales:18.1f}  {str(raining):>10}")
    
    
    
    
```

```python
import numpy as np

a1 = np.mean(temperatures)
a2 = np.mean(ice_cream_prices)
a3 = np.mean(No_of_tourists)
a4 = np.mean(ice_cream_Sales)

print("Average Temperature:", a1)
print("Average Ice Cream Prices:", a2)
print("Average Number of Tourists:", a3)
print("Average Ice Cream Sales:", a4)

a5 = np.median(temperatures)
a6 = np.median(ice_cream_prices)
a9 = np.median(No_of_tourists)
a10 = np.median(ice_cream_Sales)

print("Median Temperature:", a5)
print("Median Ice Cream Price:", a6)
print("Median Number of Tourists:", a9)
print("Median Ice Cream Sales:", a10)

a7 = np.max(temperatures)
a8 = np.max(ice_cream_prices)
a11 = np.max(No_of_tourists)
a12 = np.max(ice_cream_Sales)

print("Maximum Temperature:", a7)
print("Maximum Ice Cream Prices:", a8)
print("Maximum Number of Tourists:", a11)
print("Maximum Ice Cream Sales:", a12)

print("Standard Deviation of Temperature:", np.std(temperatures))
print("Standard Deviation of Ice Cream Prices:", np.std(ice_cream_prices))
print("Standard Deviation of Number of Tourists:", np.std(No_of_tourists))
print("Standard Deviation of Ice Cream Sales:", np.std(ice_cream_Sales))
```

```python
# Do hot days correlate with ice cream sales?

correlation = np.corrcoef(temperatures, ice_cream_prices)
print("Correlation between Temperature and Ice Cream Sales:", correlation[0, 1])
```

```python
# Are high prices reducing purchases?
correlation = np.corrcoef(ice_cream_prices, ice_cream_Sales)
print("Correlation between Ice Cream Prices and Sales:", correlation[0, 1])
```

```python
# Are tourists boosting demand?

correlation = np.corrcoef(No_of_tourists, ice_cream_Sales)
print("Correlation between Number of Tourists and Ice Cream Sales:", correlation[0, 1])
```

```python
# Are sales rising or falling daily?
daily_sales_trend = np.diff(ice_cream_Sales)
print("Daily Sales Trend:", daily_sales_trend)
```

```python
# Are sales consistent or unpredictable?

Variation = np.std(ice_cream_Sales)
print(Variation)

```

```python
# Detect which days had unusually high or low sales:

mean_sales = np.mean(ice_cream_Sales)
std_sales = np.std(ice_cream_Sales)

outliers = ice_cream_Sales[np.abs(ice_cream_Sales - mean_sales) > 2 * std_sales]
print("Outlier Sales Days:", outliers)
```
